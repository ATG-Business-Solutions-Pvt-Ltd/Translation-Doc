```python
import io
import fitz
from transformers import MBartForConditionalGeneration, MBart50Tokenizer
```

```python
model_name = "facebook/mbart-large-50-many-to-many-mmt"
model = MBartForConditionalGeneration.from_pretrained(model_name)
tokenizer = MBart50Tokenizer.from_pretrained(model_name)
```

```python
def upload_file():
    file_path = input("Enter the file path: ")
    with open(file_path, 'rb') as file:  # Read file in binary mode
        file_content = file.read()
    print("filename:", file_path.split('/')[-1])
    return file_content



def translate_pdf(file_content):
    # List of supported languages and their codes
    languages = {
    "Arabic": "ar_AR",
    "Czech": "cs_CZ",
    "German": "de_DE",
    "English": "en_XX",
    "Spanish": "es_XX",
    "Estonian": "et_EE",
    "Finnish": "fi_FI",
    "French": "fr_XX",
    "Gujarati": "gu_IN",
    "Hindi": "hi_IN",
    "Italian": "it_IT",
    "Japanese": "ja_XX",
    "Kazakh": "kk_KZ",
    "Korean": "ko_KR",
    "Lithuanian": "lt_LT",
    "Latvian": "lv_LV",
    "Burmese": "my_MM",
    "Nepali": "ne_NP",
    "Dutch": "nl_XX",
    "Romanian": "ro_RO",
    "Russian": "ru_RU",
    "Sinhala": "si_LK",
    "Turkish": "tr_TR",
    "Vietnamese": "vi_VN",
    "Chinese": "zh_CN",
    "Afrikaans": "af_ZA",
    "Azerbaijani": "az_AZ",
    "Bengali": "bn_IN",
    "Persian": "fa_IR",
    "Hebrew": "he_IL",
    "Croatian": "hr_HR",
    "Indonesian": "id_ID",
    "Georgian": "ka_GE",
    "Khmer": "km_KH",
    "Macedonian": "mk_MK",
    "Malayalam": "ml_IN",
    "Mongolian": "mn_MN",
    "Marathi": "mr_IN",
    "Polish": "pl_PL",
    "Pashto": "ps_AF",
    "Portuguese": "pt_XX",
    "Swedish": "sv_SE",
    "Swahili": "sw_KE",
    "Tamil": "ta_IN",
    "Telugu": "te_IN",
    "Thai": "th_TH",
    "Tagalog": "tl_XX",
    "Ukrainian": "uk_UA",
    "Urdu": "ur_PK",
    "Xhosa": "xh_ZA",
    "Galician": "gl_ES",
    "Slovene": "sl_SI"
}


    # Print the languages and let the user choose one
    for i, language in enumerate(languages.keys(), start=1):
        print(f"{i}. {language}")
    language_number = int(input("Enter the number of the language you want to translate to: ")) - 1
    source_language = list(languages.values())[language_number]

    from io import BytesIO
    import fitz  # PyMuPDF

    try:
        # Create a file-like object from the file content
        file_object = BytesIO(file_content)

        # Open the PDF file
        doc = fitz.open(stream=file_object, filetype='pdf')

        translations = []

        for page in doc:
            text = page.get_text()

            # Set the source language
            tokenizer.src_lang = f"{source_language}"
            encoded = tokenizer(text, return_tensors="pt")
            generated_tokens = model.generate(**encoded, forced_bos_token_id=tokenizer.lang_code_to_id["en_XX"])
            out = tokenizer.batch_decode(generated_tokens, skip_special_tokens=True)

            translations.append(str(out).strip('][\''))

        return translations
    except Exception as e:
        print(f"An error occurred while reading the PDF: {e}")
        return None



```

```python
# Use the function
FileContent = upload_file()
translations = translate_pdf(FileContent)

if translations is not None:
    for translation in translations:
        print(translation)
else:
    print("An error occurred while translating the PDF.")
```
